**Split the data in train and test**

1.   Features is the list of all the Diseases +  Symptpms. 
2.   Target is the list of all the Food.



```python
def split_data(Feature, Target, data):
  X = data.loc[:, Feature]
  Y = data.loc[:, Target]
  return X, Y 
```

**Split the data into the X_train and Y _train, X_test and Y_test**

```python
def Split_data_train_test(X, Y):
  X_train, X_test, Y_train, Y_test = train_test_split(X, Y, random_state = 42, train_size = .8)
  return X_train, X_test, Y_train, Y_test
```

**Creat the model**

```python
def model(len_diseases_and_symptoms_in_row, len_food_in_row):
  mymodel = tf.keras.Sequential([
     tf.keras.layers.Dense(len_diseases_and_symptoms_in_row),
     tf.keras.layers.Dense(600, activation = 'relu'),
     tf.keras.layers.Dense(1200,  activation = 'relu'),
     tf.keras.layers.Dense(2400, activation = 'relu'),
     tf.keras.layers.Dense(4800,activation = 'relu'),
    #  tf.keras.layers.Dense(7000, activation = 'relu'),
    #  tf.keras.layers.Dense(8000, activation = 'relu'), # elu, gelu, softmax, sigmoid
    #  tf.keras.layers.Dense(9000,activation = 'relu'),
    #  tf.keras.layers.Dense(10000),

     tf.keras.layers.Dense(4500, activation = 'relu'),
     tf.keras.layers.Dense(2200, activation = 'relu'),
     tf.keras.layers.Dense(1100, activation = 'relu'),
     tf.keras.layers.Dense(550, activation = 'relu'),
     tf.keras.layers.Dense(225, activation = 'relu'),
     tf.keras.layers.Dense(112, activation = 'relu'),
     tf.keras.layers.Dense(56, activation = 'relu'),
     #tf.keras.layers.Dense(10000),
     tf.keras.layers.Dense(len_food_in_row, activation = 'relu')
  ])
  return mymodel
```

**compile the model**

```python
def compile_my_model(model):
  model.compile(loss = tf.keras.losses.mae,
                optimizer = tf.keras.optimizers.Adam(lr =  0.0002),  
                metrics = ["mae"])
```

**See the model diagram**

```python
def model_diagram(model):
  plot_model(model = model, show_shapes = True)
```

**Fit model**

```python
def fit_model(X_train, Y_train, model, epochs):
  model.fit(X_train, Y_train, epochs)
```

**Predict the model**

```python
def predict_model(X_test, model):
  Y_pred = model.predict(X_test)
  return Y_pred
```

**Evaluate the model**

```python
def evalute_model(X_test, Y_test, model):
  model.evaluate(X_test, Y_test)
```
